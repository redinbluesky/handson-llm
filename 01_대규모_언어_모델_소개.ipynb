### Chapter 1-2-1 BoW로 언어 표현하기 <a class="anchor" id="chapter1-2-1"></a>
1. BoW(Bag of Words)는 텍스트를 표현하는 가장 간단한 방법 중 하나이다.

2. 수치 표현을 만들려는 두 개의 문장이 있다고 가정한다.
    - That is a cute dog
    - My cat is cute

3. 토큰화(tokenization)을 통해 문장을 개별 단어나 부분단인 토큰(token)으로 분리한다.
    - 가장 널리 사용되는 토큰화 방법은 공백을 기준으로 나누는것이다.
    - that is a cut dog my cat is cute

4. 고유한 단어를 합쳐 어휘 사전을 만든다.
    - 중복되는 단어는 제외한다.
    - that is a cut dog my

5. 어휘사전을 사용하여 각 문장에 단어가 얼마나 많이 등장하는지 헤아린다.
    - 아래의 그림과 같이 벡터 형태의 표현으로 텍스트가 표현된다.

        ![BoW 예시](./image/01_bow_example.png)

6. 텍스트를 벡터 또는 벡터 표현이라고 부르는 수치의 형태로 변환하는 모델을 표현 모델이라고 한다.
    - BoW는 표현 모델의 한 예이다.

### Chapter 1-2-2 밀집 벡터 임베딩으로 더 나은 표현 만들기 <a class="anchor" id="chapter1-2-2"></a>
1. BoW의 단점은 언어를 가방에 담긴 단어의 모음 이상으로 생각하지 않으며 텍스트의 의미를 무시한다.

2. word2vec은 임베딩(embedding)으로 텍스트의 의미를 포착한 첫 번째 시도였다.

3. 임베딩은 데이터의 의미를 포착하기 위한 벡터표현이다.
    - word2vec은 위키백과와 같은 방대한 텍스트에서 훈련하여 단어의 의미를 나타내는 표현을 학습했다.

4. word2vec은 신경망을 사용하여 의미있는 표현을 생성한다.
    - 이 신경망은 여러 층이 서로 연결되어 구성되며, 각 층은 정보를 처리하는 여러 노드로 이루어져 있다.
    - 한 층의 노드와 다음 층의 노드를 잇는 연결에는 입력값에 따라 특정한 가중치가 부여되는데 이를 모델 파라미터라고 부른다.

        ![word2vec 신경망 구조](./image/01_word2vec_network2.png)

5. word2vec은 신경망을 사용해 주어진 문장에서 다음에 어떤 단어가 등장하는 지를 살펴봄으로써 단어 임베딩을 생성한다.

6. 먼저 어휘사전에 있는 모든 단어에 대해서 랜덤하게 초기화된 일련의 값(사용자가 지정해야 하는 값: 하이퍼파라미터)을 지정한다.
    - 구글이 공개한 사전 훈련된 word2vec 모델의 임베딩 차원은 300차원이다.
    - 한개의 토큰이 300개의 차원을 가지고 있다.

7. 각 훈련 스텝마다 훈련 데이터에서 단어 쌍을 가져온 후 모델이 문장 안에서 단어 쌍이 이웃에 나탄날 가능성이 있는지 예측한다.

    ![word2vec 훈련 과정](./image/01_word2vec_training.png)


8. 훈련 과정 동안 word2vec은 단어 사이 관계를 학습하고 이 정보를 임베딩에 저장한다.

9. 이렇게 만들어진 임베딩은 단언의 의미를 포착한다.
    - 임베딩은 단어의 의미를 표현하기 위해 속성을 같는다.
    - 임베딩은 크기가 고정이기 때문에 임베딩 속성은 단어의 정신적 표상을 만들도록 선택된다.
    - 아래의 이미지는 속성 개념을 상상해서 만든것이다.

        ![임베딩 시각화](./image/01_embedding_visualization.png)
    
10. 이러한 속성은 매우 모호하여 하나의 객체나 사람이 식별할 수 있는 개념과 일치하지 않을 수 있다.
    - 하지만 컴퓨터에게는 의미가 있으며 사람의 언어를 컴퓨터로 바꾸는데 도움이 된다.

11. 임베딩은 단어 사이에 있는 의미 유사성을 측정할 수 있기 때문에 매우 유용하다.
    - 다양한 거리 측정 방법을 사용하여 단어 사이의 유사성을 측정할 수 있다.
