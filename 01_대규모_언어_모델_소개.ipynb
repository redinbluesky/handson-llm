### Chapter 1-2-1 BoW로 언어 표현하기 <a class="anchor" id="chapter1-2-1"></a>
1. BoW(Bag of Words)는 텍스트를 표현하는 가장 간단한 방법 중 하나이다.

2. 수치 표현을 만들려는 두 개의 문장이 있다고 가정한다.
    - That is a cute dog
    - My cat is cute

3. 토큰화(tokenization)을 통해 문장을 개별 단어나 부분단인 토큰(token)으로 분리한다.
    - 가장 널리 사용되는 토큰화 방법은 공백을 기준으로 나누는것이다.
    - that is a cut dog my cat is cute

4. 고유한 단어를 합쳐 어휘 사전을 만든다.
    - 중복되는 단어는 제외한다.
    - that is a cut dog my

5. 어휘사전을 사용하여 각 문장에 단어가 얼마나 많이 등장하는지 헤아린다.
    - 아래의 그림과 같이 벡터 형태의 표현으로 텍스트가 표현된다.

        ![BoW 예시](./image/01_bow_example.png)

6. 텍스트를 벡터 또는 벡터 표현이라고 부르는 수치의 형태로 변환하는 모델을 표현 모델이라고 한다.
    - BoW는 표현 모델의 한 예이다.

### Chapter 1-2-2 밀집 벡터 임베딩으로 더 나은 표현 만들기 <a class="anchor" id="chapter1-2-2"></a>
1. BoW의 단점은 언어를 가방에 담긴 단어의 모음 이상으로 생각하지 않으며 텍스트의 의미를 무시한다.

2. word2vec은 임베딩(embedding)으로 텍스트의 의미를 포착한 첫 번째 시도였다.

3. 임베딩은 데이터의 의미를 포착하기 위한 벡터표현이다.
    - word2vec은 위키백과와 같은 방대한 텍스트에서 훈련하여 단어의 의미를 나타내는 표현을 학습했다.

4. word2vec은 신경망을 사용하여 의미있는 표현을 생성한다.
    - 이 신경망은 여러 층이 서로 연결되어 구성되며, 각 층은 정보를 처리하는 여러 노드로 이루어져 있다.
    - 한 층의 노드와 다음 층의 노드를 잇는 연결에는 입력값에 따라 특정한 가중치가 부여되는데 이를 모델 파라미터라고 부른다.

        ![word2vec 신경망 구조](./image/01_word2vec_network2.png)

5. word2vec은 신경망을 사용해 주어진 문장에서 다음에 어떤 단어가 등장하는 지를 살펴봄으로써 단어 임베딩을 생성한다.

6. 먼저 어휘사전에 있는 모든 단어에 대해서 랜덤하게 초기화된 일련의 값(사용자가 지정해야 하는 값: 하이퍼파라미터)을 지정한다.
    - 구글이 공개한 사전 훈련된 word2vec 모델의 임베딩 차원은 300차원이다.
    - 한개의 토큰이 300개의 차원을 가지고 있다.

7. 각 훈련 스텝마다 훈련 데이터에서 단어 쌍을 가져온 후 모델이 문장 안에서 단어 쌍이 이웃에 나탄날 가능성이 있는지 예측한다.

    ![word2vec 훈련 과정](./image/01_word2vec_training.png)


8. 훈련 과정 동안 word2vec은 단어 사이 관계를 학습하고 이 정보를 임베딩에 저장한다.

9. 이렇게 만들어진 임베딩은 단언의 의미를 포착한다.
    - 임베딩은 단어의 의미를 표현하기 위해 속성을 같는다.
    - 임베딩은 크기가 고정이기 때문에 임베딩 속성은 단어의 정신적 표상을 만들도록 선택된다.
    - 아래의 이미지는 속성 개념을 상상해서 만든것이다.

        ![임베딩 시각화](./image/01_embedding_visualization.png)
    
10. 이러한 속성은 매우 모호하여 하나의 객체나 사람이 식별할 수 있는 개념과 일치하지 않을 수 있다.
    - 하지만 컴퓨터에게는 의미가 있으며 사람의 언어를 컴퓨터로 바꾸는데 도움이 된다.

11. 임베딩은 단어 사이에 있는 의미 유사성을 측정할 수 있기 때문에 매우 유용하다.
    - 다양한 거리 측정 방법을 사용하여 단어 사이의 유사성을 측정할 수 있다.
    - 아래의 그림처럼 임베딩을 2차원으로 표현으로 압축하면 비슷한 의미의 단어가 서로 가깝게 위치하는 것을 볼 수 있다.

        ![임베딩 시각화](./image/01_embedding_visualization3.png)

### Chapter 1-2-3 임베딩의 종류 <a class="anchor" id="chapter1-2-3"></a>
1. 임베딩의 종류는 많이 있으며, 단어 임베딩과 문장 임베딩은 다른 수준의 추상화에 사용된다.
    - BoW는 전체 문서를 표현하므로 문장 임베딩의 한 형태로 간주될 수 있다.
    - word2vec은 단어에 대한 임베딩만 생성한다.
    
        ![임베딩 종류](./image/01_embedding_types.png)

### Chapter 1-2-4 어텐션을 사용한 문맥 인코딩과 디코딩 <a class="anchor" id="chapter1-2-4"></a>
1.  word2vec은 정적인 단어 표현을 만든다.
    - 예를 들어 단어 'bank'는 문맥에 상관없이 항상 임베딩이 동일하다.
    - 은행, 강뚝과 같이 문맥에 따라 의미가 달라져야 하는 단어의 경우 word2vec은 문맥을 포착할 수 없다.
    
2. 이 문제는 RNN을 사용하여 해결되었다.

3. 영어에서 네덜란드어로 번역하는 RNN 모델을 아래의 이미지와 같다고 생각해보자.

    ![RNN 번역 모델](./image/01_rnn_translation.png)

4. RNN은 각 단계에서 이전에 생성한 모든 단어를 사용해 다음 단어를 생성한다.

    ![RNN 번역 모델](./image/01_rnn_translation2.png)


5. 인코딩 단계의 목표는 입력을 가능한 잘 표현하여 디코더의 입력으로 사용되는 임베딩의 형태로 문맥을 생성하는 것이다.
    - 문맥은 토큰으로 구성되기 때문에 초기 표현으로 word2vec과 같은 단어 임베딩을 사용할 수 있다.

        ![RNN 인코딩 단계](./image/01_rnn_encoding.png)

6. 문맥 임베딩은 하나의 임베딩으로 전체 입력을 표현하기 때문에 긴 문장을 처리하기 어렵다.

7. 이 문제는 어텐션(attention) 메커니즘을 사용하여 해결되었다.
    - 입력 시퀸스에서 서로 관련 있는 부분에 모델이 초점을 맞추고 해당 신호를 증폭한다.
    - 어텐션은 주어진 문장에서 어떤 단어가 가장 중요한지 모델이 학습할 수 있도록 한다.

8. 예를 들어 출력단어 "lama's"는  입력단어 "llamas"에 해당하는 네덜란드어 이기 때문에 두  단어 사이의 어텐션은 높다.

9. 예제의 입출력 사이의 어텐션 맵은 아래의 이미지와 같다.

    ![RNN 어텐션 맵](./image/01_rnn_attention_map.png)

10. 어텐션 매커니즘을 디코더 단계에 추가해서 RNN이 입력 시퀸스의 모든 단어에 접근할 수 있도록 하면 긴 문장도 처리할 수 있다.
    - 문맥 임베딩을 전달하는 대신 모든 입력 단어의 은닉 표현을 디코더에 전달한다.
    
11. RNN이 "I love llamas"라는 문정을 번역할 때 "llamas"에 대한 어텐션이 "lama's"에 집중되는 것을 볼 수 있다.

12. word2vec과 비교하면 이 구조는 텍스트의 순차적 특징과 전체 문장에 대한 이해를 모두 포착할 수 있다.

    - 이런 순차적 특징은 문장을 병렬화 하는데 방해가 된다.

### Chapter 1-2-5 Attentions Is All You Need <a class="anchor" id="chapter1-2-5"></a>
1. 2017년 "Attentions Is All You Need" 논문에서 어텐션 메커니즘만 사용하고 순환 신경망을 제거한 트랜스포머라는 신경망 구조를 제안했다.

2. 트래스포머의 인코더와 디코더는 아래의 그림과 같이 차고차곡 싸여있다.
    - 트랜스포머도 자귀회귀적이여서 이전에 생성된 모든 단어를 사용해 새로운 단어를 생성한다.
    - 인코더와 디코더 블록은 어텐션 기능이 포함된 RNN을 활용하는 데신 어텐션을 중심으로 구동된다.

        ![트랜스포머 구조](./image/01_transformer_structure.png)

3. 트랜스포머의 인코더 블록은 셀프 어텐션과 피드포원드 신경망 두 부분으로 구성된다.
    - 셀프 어텐션은 한 시퀸스 안의 다른 위치에 주의를 기울인다.
    
        ![트랜스포머 인코더 블록](./image/01_transformer_encoder_block.png)

4. 입력 시퀸스를 더 쉽고 정확하게 표현할 수 있도록 어텐션 매커니즘이 개선되었다.
    - 한 번에 하나의 토큰을 처리하지 않고 전체 시퀸스를 한번에 처리한다.

    ![RNN 어텐션 맵](./image/01_rnn_attention_map.png)

5. 디코더는 인코더의 출력에 주의를 기울이는 별도의 층이 추가된다.

    ![트랜스포머 디코더 블록](./image/01_transformer_decoder_block.png)

6. 디코더의 셀프 어텐션 층은 미새의 위치를 마스킹한다.
    - 디코더가 다음 단어를 예측할 때 이전 단어에만 주의를 기울이도록 한다.

### Chapter 1-2-6 표현모델:인코더 기반 모델<a class="anchor" id="chapter1-2-6"></a>
1. 원본 트랜스포머 모델은 인코더-디코더 구조라서 번역 작업에는 잘 맞지만 텍스트 분류와 같은 작업에는 최적이 아니다.

2. 2018년 BERT 모델이 제안되면서 트랜스포머의 인코더 부분만 사용하는 새로운 모델 구조가 등장했다.
    - BERT는 언어를 표현하는 데 초점을 맞춘 인코더 기반 모델이다.
    
        ![BERT 구조](./image/01_bert_structure.png)

3. BERT는 마스크드 언어 모델링 기법을 적용하였다.
    - 모델이 예측할 입력의 일부분을 마스킹 하여 모델이 문맥을 이해하도록 훈련한다.

        ![BERT 마스크드 언어 모델링](./image/01_bert_masked_language_modeling.png)

4. BERT와 탐은 모델은 일반적으로 전이학습(transfer learning)에 사용된다.
    - 언어 모델링을 위해 모델을 사전 훈련하고 그다음 구체적인 작업을 위해 미세튜닝을 한다.
    - 예를 들어, 위키백과 전체로 BERT 모델을 훈련하여 단어의 의미를 포착하는 임베딩을 학습한 다음, 이 모델을 사용하여 감정 분석과 같은 특정 작업에 맞게 미세튜닝할 수 있다.

5. 사전 훈련 모델의 장점은 대부분의 훈련이 이미 완료되었다는 것이다.
    - 특정 작업을 위한 미세튜닝은 일반적으로 컴퓨팅 자원과 데이터를 적게 사용한다.

6. BERT와 같은 모델은 거의 모든 단계에서 임베딩을 생성한다.
    - 출력크기가 단어 임베딩 크기와 동일한 임베딩을 생성한다.

7. 이책에서 인코더 기반 모델을 표현 모델이라고 부르고, 디코더 기반 모델을 생성 모델이라고 부르겠다.
    - 표현 모델은 언어를 표현하는 데 초점을 맞추고 텍스트를 생성하지 않는다.
    - 생성 모델은 주로 텍스트를 생성하는데 초점을 맞추고, 임베딩을 생성하도록 훈련되지 않는다.

2. 대규모 언어 모델의 정의는 새로운 모델이 출시됨에 따라 진화한다.

## Chapter 1-4 대규모 언어 모델의 훈련 패러다임<a class="anchor" id="chapter1-4"></a>
1. 전통적인 머신러닝은 분류와 같은 특정 작업을 위해 모델을 훈련한다.
    - 데이터(주로 구조적인 데이터)에서 훈련 후 모델이 새로운 데이터에 대한 예측을 할 수 있도록 한다.

2. LLM은 일반적으로 두 단계로 훈련된다.
    - 언어 모델링과 미세튜닝이다.
    - 언어 모델링에서는 모델이 대규모 텍스트 데이터에서 언어의 패턴과 구조를 학습한다.
        - 훈련 결과 다음 단어를 예측하는 모델이 만들어지며 이렇게 만들어진 모델을 사전 훈련된 모델, 파운데이션 모델이라고 부른다.
    - 미세튜닝에서는 언어 모델링으로 학습된 모델을 특정 작업이나 도메인에 맞게 조정한다.

## Chapter 1-9 첫 번째 텍스트 생성하기<a class="anchor" id="chapter1-9"></a>
1. 어떤 언어 모델을 사용할지 선택하는 것은 중요한 일이다.
    - LLM을 찾아보고 다운로드 하기 좋은 곳은 허깅 페이스 허브이다.
    - 허깅 페이스는 유명한 트랜스포머스 패키지를 만든 회사이다.

2. 생성 모델로 주로 Phi-3-mini를 사용한다.
    - 비교적 작은 38억개의 파라미터로 GPT-3보다 훨씬 작은 모델이지만 GPT-3와 유사한 성능을 보여준다.

3. LLM을 사용하기 위해 생성모델 자체와 토크나이져를 로드한다.
    - 토크나이져는 텍스트를 모델이 이해할 수 있는 토큰으로 변환하는 도구이다.
    - 허깅페이스 사이트 주소에 모델에 해당하는 ID를 연결하면 토크나이져와 모델을 찾을 수 있다.

```python
# 깃허브에서 위젯 상태 오류를 피하기 위해 진행 표시줄을 나타내지 않도록 설정합니다.
from transformers.utils import logging

logging.disable_progress_bar()
```

```python
# 트랜스포머스에서 "microsoft/Phi-3-mini-4k-instruct" 모델과 토크나이져 로드
from transformers import AutoTokenizer, AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",      # 모델이 GPU에서 실행되도록 설정
    torch_dtype="auto",     # 모델이 자동으로 최적의 데이터 타입을 선택하도록 설정
    trust_remote_code=False, # 모델이 사용자 정의 코드를 포함하는 경우 이를 신뢰하도록 설정 
)

tokenizer = AutoTokenizer.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    trust_remote_code=True, # 모델이 사용자 정의 코드를 포함하는 경우 이를 신뢰하도록 설정
)
```

```python
from transformers import pipeline

# 파이프라인을 만듭니다.
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False
)
```

```python
# 프롬프트 (사용자 입력 / 쿼리)
messages = [
    {"role": "user", "content": "Create a funny joke about chickens."}
]

# 출력 생성
output = generator(messages)
print(output[0]["generated_text"])
```
